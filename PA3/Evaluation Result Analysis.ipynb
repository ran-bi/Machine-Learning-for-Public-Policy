```python
import pandas as pd
import numpy as np
```

```python
result = pd.read_csv('evaluation/clf_evaluations.csv', index_col=0)
METRICS = ['auc', 'f1_score','precision','recall','accuracy','train_time']
```

```python
def select_best(result, metric):
    best_models_by_metric = result.sort(metric, ascending=False).groupby('model_type', as_index=False).first().sort(metric, ascending=False)
    return best_models_by_metric
```

```python
def best_all(result, metrics):
    best_models = {}
    for metric in metrics:
        best_models[metric] = select_best(result, metric)
    return best_models
        
```

```python
best_models = best_all(result, METRICS)
```
